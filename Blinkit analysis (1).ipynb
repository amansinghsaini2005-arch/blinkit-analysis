## **BUSINESS REQUIRMENT**

### **KPI'S REQUIRMENT**

```python
# Total Sales
total_sales = df['Sales'].sum()

# Average Sales
avg_sales = df['Sales'].mean()

# No of Items Sold
no_of_items_sold = df['Sales'].count()

# Average Rating
avg_rating = df['Rating'].mean()

# Display

print(f"Total Sales: ${total_sales:,.0f}")
print(f"Average Sales: ${avg_sales:,.0f}")
print(f"No of Items sold: {no_of_items_sold:,.0f}")
print(f"Average Rating: {avg_rating:,.1f}")
```

### **CHARTS REQUIRMENT**

#### **Total Sales by Fat Content**

```python
# Grouping sales by fat content
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

# Pie chart
plt.pie(
    sales_by_fat, 
    labels=sales_by_fat.index,
    autopct='%.1f%%',
    startangle=90
)
plt.title('Sales by Fat Content')
plt.axis('equal')  # Make the pie chart circular
plt.show()
```

#### **Total Sales by Items Type**

```python
# Group data by 'Item Type' and calculate total sales for each type
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

# Create a figure of size 10x6 inches
plt.figure(figsize=(10,6))

# Create a bar chart with item types on X-axis and sales on Y-axis
bars = plt.bar(sales_by_type.index, sales_by_type.values)

# Rotate X-axis labels for better visibility
plt.xticks(rotation=-90) 

# Label the X and Y axes
plt.xlabel('Item Type')   # Label for X-axis
plt.ylabel('Total Sales') # Label for Y-axis

# Add a title to the chart
plt.title('Total Sales by Item Type')

# Add text labels above each bar showing the exact sales value
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,   # X position (middle of the bar)
        bar.get_height(),                    # Y position (top of the bar)
        f'{bar.get_height():,.0f}',          # Text (formatted sales value)
        ha='center',                         # Horizontal alignment
        va='bottom',                         # Vertical alignment
        fontsize=9                           # Font size
    )

# Show the chart
plt.tight_layout()
plt.show()


```

#### **Fat Content by Outlet for Total Sales**

```python
# Group sales by Outlet Location Type and Item Fat Content, then unstack to make columns
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()

# Keep only 'Regular' and 'Low Fat' columns (if they exist)
grouped = grouped[['Regular', 'Low Fat']]

# Create a bar chart
ax = grouped.plot(
    kind='bar',
    figsize=(8, 5),
    title='Outlet Tier by Item Fat Content'
)

# Add labels and legend
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')

# Adjust layout and show
plt.tight_layout()
plt.show()
```

#### **Total Sales by Outlet Establishment**

```python

# Group total sales by outlet establishment year
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

# Plot line chart
plt.figure(figsize=(9,5))
plt.plot(
    sales_by_year.index,
    sales_by_year.values,
    marker='o',
    linestyle='-',
    color='teal',
    linewidth=2
)

# Add grid and labels
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment Year')

# Add value labels
for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=8)

# Rotate X-axis labels slightly for better readability
plt.xticks(rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```

#### **Sales by Outlet size**

```python
# Group sales by outlet size
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

# Pie chart
plt.figure(figsize=(4,4))
plt.pie(
    sales_by_size,
    labels=sales_by_size.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Sales Distribution by Outlet Size')
plt.tight_layout()
plt.show()
```

#### **Sales by Outlated Location**

```python
# Group total sales by outlet location type
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

# Horizontal bar chart
plt.figure(figsize=(8,5))
plt.barh(sales_by_location['Outlet Location Type'], sales_by_location['Sales'], color='skyblue')

plt.title('Total Sales by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

# Add data labels
for index, value in enumerate(sales_by_location['Sales']):
    plt.text(value, index, f'{value:,.0f}', va='center', fontsize=8)

plt.tight_layout()  # Ensures no cutoff
plt.show()
```
